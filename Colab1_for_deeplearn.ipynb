# "Hello World" нейронных сетей

Как и в каждом первом приложении, вы должны начать с чего-то очень простого, показывающего общую структуру работы вашего кода.

В случае создания нейронных сетей, я хотел бы использовать пример, где он изучает отношения между двумя числами. Так, например, если вы писали код для такой функции, вы уже знаете «правила» -


```
float hw_function(float x){
    float y = (2 * x) - 1;
    return y;
}
```

Итак, как бы вы обучили нейронную сеть выполнять аналогичные задачи? Используя данные! Наполнив ее набором пар X и Y, вы обучите ее отношению между ними.

Это, очевидно, совсем другая парадигма, чем та, к которой вы, возможно, привыкли, поэтому давайте рассмотрим ее по частям.


## Импорты внешних библиотек

Начнем с импорта. Здесь мы импортируем **TensorFlow** и назовем его **tf** для простоты использования.

Затем мы импортируем библиотеку **numpy**, которая помогает нам легко и быстро представлять наши данные в виде списков.

Фреймворк для определения нейронной сети как набора последовательных слоев называется **keras**, и мы его также импортируем.

````
import tensorflow as tf
import numpy as np
from tensorflow import keras
````

## Определить и скомпилировать нейронную сеть

Далее мы создадим простейшую из возможных нейронных сетей. У этого есть 1 слой, у этого слоя есть 1 нейрон, и входная форма к нему - только 1 значение.

````
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
````

Теперь компилируем нашу нейронную сеть. Надо указать 2 функции: loss и optimizer.
вот здесь нужна математика, которую вы уже рассматривали в предыдущих лекциях, но в этом случае она уже хорошо упакована в функции для вас.
Мы знаем, что в нашей функции соотношение между числами равно y = 2x-1.

Когда компьютер пытается «узнать» эту функцию, он делает предположение: возможно, у = 10х + 10. Функция LOSS сравнивает предполагаемые ответы с известными правильными ответами и измеряет, насколько хорошо или плохо всё сработало.
Затем он использует функцию OPTIMIZER, чтобы сделать другое предположение. Основываясь на том, как сработала функция потерь, она попытается минимизировать потери. В этот момент, может быть, получится что-то вроде y = 5x + 5, которое, хотя все еще довольно плохо, ближе к правильному результату (то есть потери стали меньше)
Это будет повторяться для количества EPOCHS, которое вы увидите в ближайшее время. Но сначала, вот как мы говорим ему использовать «MEAN SQUARED ERROR» для функции потерь и «STOCHASTIC GRADIENT DESCENT» для оптимизатора. И вы уже знаете математику, которая в них работает
Со временем вы освоите разные функции потерь и оптимизатора, подходящие для разных сценариев.

````
model.summary()
````

````
model.compile(optimizer='sgd', loss='mean_squared_error')
````

## Предоставление данных

Далее мы введем некоторые данные. В этом случае мы берем 6 иксов и 6 игреков. Вы можете видеть, что отношения между ними таковы: y = 2x-1, где x = -1, y = -3 и т. д.
Библиотека Python под названием «numpy» предоставляет множество структур данных типа массива, которые являются стандартным способом де-факто. Мы заявляем, что хотим использовать их, указав значения как np.array []

````
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
````

# Обучение нейронной сети

Процесс обучения нейронной сети, где она «изучает» отношения между X и Y, происходит в вызове ** model.fit **. Здесь мы пройдем через цикл, о котором мы говорили выше, сделаем предположение, измерим, насколько оно хорошо или плохо (каковы потери), используя opimizer, чтобы сделать другое предположение и т.д. Функция model.fit сделает это за количество эпох, которое вы укажете. Когда вы запустите этот код, справа вы увидите потери.

````
model.fit(xs, ys, epochs=500)
````

Хорошо, теперь у вас есть модель, которая была обучена отношениям между X и Y. Вы можете использовать метод **model.predict**, чтобы он вычислил Y для ранее неизвестного X. Так, например, Если X = 10, как вы думаете,какой  будет Y? Сделайте предположение, прежде чем запускать этот код:

````
print(model.predict([10.0]))
````

Вы могли подумать 19, верно? Но оно оказалось немного меньше. Как вы думаете, почему?
Помните, что нейронные сети имеют дело с вероятностями, поэтому, на основании данных, которыми мы снабдили нейронную сеть, было рассчитано, что существует очень высокая вероятность того, что соотношение между X и Y будет $$Y = 2X-1$$ но имея только 6 точек данных, мы не можем знать наверняка. В результате значение для 10 очень близко к 19, но не обязательно 19.
Работая с нейронными сетями, вы увидите, что почти всегда будете иметь дело с вероятностями, а не с определенностью, особенно когда речь идет о классификации.
